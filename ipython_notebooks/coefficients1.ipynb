```python
s = 9

a = [[-1 for j in range(s+2)] for i in range(s+2)]
alpha =[[-1 for j in range(s+2)] for i in range(s+2)]
c = [-1 for i in range(s+2)]
b = [-1 for i in range(s+2)]
b_hat = [-1 for i in range(s+2)]
beta = [-1 for i in range(s+2)]
beta_hat = [-1 for i in range(s+2)]

for i in range(s+2):
    for j in range(s+2):
        alpha[i][j] = 'alpha' + str(i) + str(j) # Assume i and j not > 9.
        a[i][j] = 'a' + str(i) + str(j)
    c[i] = 'c' + str(i)
    b[i] = 'b' + str(i)
    b_hat[i] = 'b_hat' + str(i)
    beta[i] = 'beta' + str(i)
    beta_hat[i] = 'beta_hat' + str(i)
```

```python
Vdot = [-1 for i in range(s+2)]
for j in range(s+2):
    Vdot[j] = 'Vdot' + str(j) + '[k+3]'
Q = 'ytemp[k]'
V = 'ytemp[k + 3]'
```

```python
# think of Q[i][k] as "ytemp[k]"
# and V[i][k] as "ytemp[k + 3]"
# and Vdot[j] as "Vdot_j[k+3]", this is so we don't have to unpack dydx[3..5]
# to get the Vdot values
Instructions = []

#duplicate_stage = [3, 6, 8]
duplicate_stage = []


for i in range(2, s + 1):
    using_last_B_field_evaluation = False
    #
    
    if not i in duplicate_stage:
        for_loop_expr = 'for (k = 0; k < 3; k ++) {'
        Instructions += [for_loop_expr]
        expression = alpha[i][1] + '*' + Vdot[1]
        for j in range(2,i):
            expression = expression + '+' + alpha[i][j] + '*' + Vdot[j]
        Q_terms = ['q0[k]', 'h*' + c[i] + '*v0[k]', 'h*h*(' + expression + ')']
        compute_expression = Q + ' = ' + Q_terms[0]
        for x in Q_terms[1:]:
            compute_expression = compute_expression + ' + ' + x
        compute_expression += ';'
        Instructions += ['\t' + compute_expression]
        end_for_loop_expr = '}'
        Instructions += [end_for_loop_expr]
    else:
        Instructions += ["// Duplicate!: using last Q_i value"] # Insert a comment
        using_last_B_field_evaluation = True  
    #
    for_loop_expr = 'for (k = 0; k < 3; k ++) {'
    Instructions += [for_loop_expr]
    expression = a[i][1] + '*' + Vdot[1]
    for j in range(2,i):
        expression = expression + ' + ' + a[i][j] + '*' + Vdot[j]
    V_terms = ['v0[k]', 'h*(' + expression + ')']
    compute_expression = V + ' = ' + V_terms[0]
    for x in V_terms[1:]:
        compute_expression = compute_expression + '+' + x
    compute_expression += ';'
    Instructions += ['\t' + compute_expression]
    end_for_loop_expr = '}'
    Instructions += [end_for_loop_expr]
    #
    # Now compute Vdot[i]:
    if using_last_B_field_evaluation == True:
        Instructions += ["// Use last B field Evaluation"]
        #Instructions += ["// Whatever we do to get the previous B field"]
        #Instruction = 'B = myField -> get_last_B_eval();'
        #Instructions += [Instruction]
        Instruction = 'ComputeRhsWithStoredB(ytemp, Vdot' + str(i) + ');'
        Instructions += [Instruction]
    else:
        RHS_Eval = 'ComputeRightHandSide(ytemp, Vdot' + str(i) +');'
        Instructions += [RHS_Eval]
        
        
        
```
